```python
# for multiple stocks 
import datetime as dt
import pandas as pd
import yfinance as yf

stocks=["MSFT","INFY.NS","0005.HK"] #0005.HK is HSBC Holdings on HK exchange
today=dt.datetime.today() 
start= today-dt.timedelta(3650) # 10 years

adj_close_prices= pd.DataFrame()
for ticker in stocks:
    adj_close_prices[ticker]=yf.download(tickers=ticker,start=start,end=today)["Adj Close"] #taking only the adjusted close value. 
```

```python
adj_close_prices.head()
```

```python
nan_counts = adj_close_prices.isna().sum()
print("Number of NaN values in each column:")
print(nan_counts)
```

```python
# using fillna to fill NaN values

#adj_close_prices.fillna(method="bfill",axis=1)
adj_close_prices.bfill(axis=1) 
# methods can be backfill, forwardfill, axis=1 means the column
```

```python
#dropna
adj_close_prices.dropna(axis=0)
```

```python
# for multiple stocks 
import datetime as dt
import pandas as pd
stocks=["MSFT","INFY.NS","0005.HK"] #0005.HK is HSBC Holdings on HK exchange
today=dt.datetime.today() 
start= today-dt.timedelta(30)

prices= pd.DataFrame()
for ticker in stocks:
    prices[ticker]=yf.download(tickers=ticker,start=start,end=today)["Adj Close"] #taking only the adjusted close value. 
```

```python
prices.head()
```

```python
#for storing the entire data we use dictionaries
all_Data= {}
for ticker in stocks:
    all_Data[ticker]=yf.download(tickers=ticker,start=start,end=today)
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
ALPHA_VANTAGE_KEY = os.getenv('ALPHA_VANTAGE_KEY')
```

```python
from alpha_vantage.timeseries import TimeSeries
import pandas as pd
ts = TimeSeries(key=ALPHA_VANTAGE_KEY,output_format='pandas')
```

```python
data = ts.get_daily(symbol='MSFT',outputsize="full")
# data is a tuple with data[0] as the pandas df and data[1] as the metadata
data[1]
```

```python
df_data=data[0]
df_data.columns=["open","high","low","close","volume"]
```

```python
df_data.head()
```

```python
# extracting ohlcv data for multiple stocks
import time 
all_tickers = ["MSFT","AMZN","GOOG"]
ohlv_dict = {}
api_call_count = 1

start_time = time.time()
for ticker in all_tickers:
    data = ts.get_intraday(symbol=ticker,interval='1min', outputsize='compact')[0]
    api_call_count+=1
    data.columns = ["open","high","low","close","volume"]
    data = data.iloc[::-1]
    ohlv_dict[ticker] = data
    if api_call_count==5:
        api_call_count = 1
        time.sleep(60 - ((time.time() - start_time) % 60.0))
```
